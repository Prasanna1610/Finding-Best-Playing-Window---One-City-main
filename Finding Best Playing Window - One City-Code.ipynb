```python
def my_resample(mydf = humid_df, freq = 'Y', mycity = 'Miami', attr = 'HUMID'):
    #ser = test_col_access(mycol)
    #z = ser.resample('Y').min()
    z = pd.Series(mydf[mycity].resample(freq).min(), name = mycity + '_' + attr + '_MIN')
    if(attr == 'TEMP'):
        #maxi = pd.Series(mydf[mycity].resample(freq).max(), name = mycity + '_' + attr + '_MAX')
        return pd.concat([mydf[mycity].resample(freq).min(), mydf[mycity].resample(freq).max()], axis = 1, \
                        keys = [mycity + '_' + attr + '_MIN', mycity + '_' + attr + '_MAX'])
    dz = pd.DataFrame(z)
    return dz

#my_resample(humid_df, freq = 'Y', mycity = 'Chicago')
```

```python
def get_filtered_rows(df = humid_df, mycity = 'Miami', start_date = '2017-01-01', end_date = '2025-01-01'):
    end_date = start_date if end_date == '2025-01-01' else end_date
    res = pd.DataFrame(df[(df.index.hour >= 5) & (df.index.hour < 22)][mycity].loc[start_date:end_date])
    return res
```

```python
def get_city_data(mycity = 'Miami', start_date = '2017-04-01', end_date = '2025-01-01'):
    
    city_humid_hr = get_filtered_rows(df = humid_df, mycity = mycity, start_date = start_date, end_date = end_date)

    city_humid_agg = my_resample(mydf = city_humid_hr, freq = 'D', mycity = mycity, attr = 'HUMID')

    city_wind_hr = get_filtered_rows(df = wind_df, mycity = mycity, start_date = start_date, end_date = end_date)

    city_wind_agg = my_resample(mydf = city_wind_hr, freq = 'D', mycity = mycity, attr = 'WIND')

    city_temp_hr = get_filtered_rows(df = temper_df, mycity = mycity, start_date = start_date, end_date = end_date)

    city_temp_agg = my_resample(mydf = city_temp_hr, freq = 'D', mycity = mycity, attr = 'TEMP')

    city_all_attr_hr = pd.concat([city_humid_hr,city_wind_hr, city_temp_hr], axis = 1)

    city_all_attr_agg = pd.concat([city_humid_agg,city_wind_agg, city_temp_agg], axis = 1)
    city_all_attr_hr['datepart'] = pd.to_datetime(city_all_attr_hr.index.date) 
    city_all_attr_hr.columns.values[[0, 1, 2]] = ['humidity', 'wind', 'temp']
    city_all_df = pd.merge(city_all_attr_hr, city_all_attr_agg, how = 'left', left_on=['datepart'], \
                            right_on=['datetime'], right_index = True).drop('datepart', axis = 1)
    city_all_df['total_var'] = city_all_df['wind'] - city_all_df[mycity+'_WIND_MIN'] + \
    city_all_df['temp'] - city_all_df[mycity+'_TEMP_MIN'] + \
    city_all_df['humidity'] - city_all_df[mycity+'_HUMID_MIN']
    city_all_df['PlayabilityRank'] = city_all_df.groupby(city_all_df.index.date)['total_var'].rank()
    return city_all_df
```

```python
final_ranked_df = get_city_data(mycity = 'Denver', start_date = '2017-07-01', end_date = '2017-07-03')
```
